```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
x = np.random.randint(low=1, high=10, size=10)
x.sort()
y = np.random.randint(low=1, high=10, size=10)
y.sort()
z = np.random.randint(low=1, high=10, size=10)
z.sort()
z = z[::-1]
print(x, y, z)
```

mfc: marker fill color <br>
marker: marker type<br>
'o': circle<br>
'--': dotted line<br>
'r': red<br>

## Line Plot

```python
plt.plot(x, y, '--', marker='o', mfc = 'r')
plt.plot(x,z, '-.', color= 'r', marker='*', mfc = 'r')
plt.plot(y,z, ':', color= 'g', marker='H', mfc = 'r')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('graph')
plt.grid()
```

## Subplotting Graphs in one Figure

```python
plt.figure(figsize = (10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y, '--', marker='o', mfc = 'r')
plt.title('subplot1')

plt.subplot(1, 2, 2)
plt.plot(x,z, '-.', color= 'r', marker='*', mfc = 'r')
plt.title('subplot2')

plt.grid()
```

## Bar Graph

```python
courses = np.array(['sjsd', 'asdsa', 'safdas', 'safda'])
students = np.random.randint(low=50, high=100, size=4)

plt.bar(courses, students, color='black', edgecolor = 'red', linewidth=1)
```

# Scatterplot

```python
plt.scatter(x, y, marker='x')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('graph')
plt.grid()
```

## Pie-Chart

```python
plt.title('Pie-Chart')
plt.pie(students, labels=courses, autopct='%.2f%%', shadow=True, explode=[0, 0, 0, 0.1])
plt.grid()
```

## Multi subplots

```python
plt.figure(figsize = (15, 5))

plt.subplot(1, 4, 1)
plt.plot(x, y, '--', marker='o', mfc = 'r')
plt.plot(x,z, '-.', color= 'r', marker='*', mfc = 'r')
plt.plot(y,z, ':', color= 'g', marker='H', mfc = 'r')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('linegraph')
plt.grid()

plt.subplot(1, 4, 2)
plt.bar(courses, students, color='black', edgecolor = 'red', linewidth=1)
plt.title('bargraph')

plt.subplot(1, 4, 3)
plt.scatter(x, y, marker='x')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('scatterplot')
plt.grid()

plt.subplot(1, 4, 4)
plt.pie(students, labels=courses, autopct='%.2f%%', shadow=True, explode=[0, 0, 0, 0.1])
plt.grid()
plt.title('piechart')

plt.subplots_adjust(wspace=0.2)
plt.grid()

plt.savefig('testingimage')
```
