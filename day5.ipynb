# Mean Imputation/Central Tendency Imputation

```python
#filling na valuse with mean of the column
df.age = df.age.fillna(df.age.mean())
df.marks = df.marks.fillna(df.marks.mode())

#since class column is non numerical we have to specify it in ['']
#mode is returned as list due to multiple "modality" hence mode()[0]
df['class'] = df['class'].fillna(df['class'].mode()[0])

df
```

```python
#missing values replaced with most recent non-missing value before them
df.ffill()
df
```

```python
df.bfill()
df
```

# Outlier Detection

```python
salry = np.random.randint(10000, 40000, 100)
salry
```

```python
q1 = np.percentile(df['salary'], 25)
q3 = np.percentile(df['salary'], 75)

q1, q3
```

```python
iqr = q3-q1
upper_fence = q3 + 1.5 * iqr
lower_fence = q1 - 1.5 * iqr
upper_fence, lower_fence
```

```python
dfwoo = df[ (df['salary'] < upper_fence) & (df['salary'] > lower_fence) ]
dfwoo
```
